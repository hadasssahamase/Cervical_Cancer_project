## Handle missing values
we identify the missing values and deal with them accordingly

```python
#create a function to calculate missing values and their percentage
def missing_values_percentage(df):

    # Total rows
    total_records = len(df)

    # Missing values
    missing_values = df.isnull().sum()

    # Percent missing values
    percent_missing_values = (missing_values/ total_records) * 100

    # # Get the data types of each column
    data_types = df.dtypes

    # create summary dataframe
    missing_values_df = pd.DataFrame({
         'missing_values': missing_values,
         'percentage (%)': percent_missing_values,
         'data_type': data_types
    })

    print(f"Total records: {total_records}")

    return missing_values_df

# View the dataframe
missing_summary =missing_values_percentage(df)
missing_summary
```

We'd drop anything with a percentage of 45 and above

```python
#filter columns with more than 45% missing percentage
cols_to_drop=missing_summary[missing_summary['percentage (%)']>45].index
#drop the filtered columns from the original dataframe
df=df.drop(columns=cols_to_drop)
#check head
df.head(10)
```

```python
#review the missing summary dataframe
missing_summary=missing_values_percentage(df)
missing_summary
```

So as we view the dataset we see an irregularity that doesn't make sense. In some rows the age when the patient had their first sexual encounter(first_sexual_intercourse) is actually greater that the current age of the patient as shown below.  

```python
#ensure both columns are numeric
df['first_sexual_intercourse']=pd.to_numeric(df['first_sexual_intercourse'],errors='coerce')
df['age']=pd.to_numeric(df['age'],errors='coerce')
#Filter where first_sexual_intercourse > age
invalid_rows=df[df['first_sexual_intercourse']>df['age']]
#display those rows
invalid_rows
```

```python
# Drop rows where first_sexual_intercourse > age
df = df[df['first_sexual_intercourse'] <= df['age']]
```

```python
# Show all rows
pd.set_option('display.max_rows', None)
#view all the rows containing missing values
missing_rows=df[df.isnull().any(axis=1)]
#view missing_rows dataframe
missing_rows
```

Since this is sensitive data it'll be wise to drop all the null values

```python
#drop null values
df=df.dropna()
#view the missing summary dataset
missing_summary=missing_values_percentage(df)
missing_summary
```

## Duplicates
we need to identify duplicates and remove them

```python
#check for duplicates
#checking for duplicates
dup=df.duplicated().sum()
if dup > 0:
    print(f'\nThere are {dup} duplicates present in our data.')
else:
    print('There are no duplicates in our data.')
```

```python
#remove duplicates
df=df.drop_duplicates()

#checking for duplicates
dup=df.duplicated().sum()
if dup > 0:
    print(f'\nThere are {dup} duplicates present in our data.')
else:
    print('There are no duplicates in our data.')
```
