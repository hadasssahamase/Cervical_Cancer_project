## Manipulating column names for better readability

```python
#check column names
df.columns
```

```python
#change column names to lowercase
df.columns=df.columns.str.lower()
#check column names
df.columns
```

```python
#replace the " " in column names with _
df.columns=df.columns.str.replace(' ','_')
#check column names
df.columns
```

```python
#replace ? with Nan
df.replace('?',np.nan,inplace=True)
#view the dataset
df.head(10)
```
